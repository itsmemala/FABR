```python
import numpy as np
from matplotlib import pyplot as plt
```

## Multi-Head

```python
path1 = 'D:\\2110465-PhD Data (DO NOT DELETE)\\hwu-MultiHead\\hwuMH-EWC.1Freeze-v17.2\\hwuMH-EWC.1Freeze-v17.2\\hwu64_bert_adapter_ewc_freeze_random'
path2 = 'D:\\2110465-PhD Data (DO NOT DELETE)\\hwu-MultiHead\\hwuMH-EWC.1\\hwuMH-EWC.1\\hwu64_bert_adapter_ewc_random'
for seed_idx in [0]:
    for rand_idx in [0]:
        list_of_lists = []
        with open(path1+str(rand_idx)+'_seed'+str(seed_idx)+'_f1.txt', 'r') as f:
            for line in f:
                inner_list = [float(elt.strip()) for elt in line.split('\t')]
                list_of_lists.append(inner_list)
            f1_matrix = np.array(list_of_lists)
            for i in range(5):
                plt.plot(list(range(i,6,1)),f1_matrix[i:,i],label='Task'+str(i)+' (LA EWC)')
for seed_idx in [0]:
    for rand_idx in [0]:
        list_of_lists = []
        with open(path2+str(rand_idx)+'_seed'+str(seed_idx)+'_f1.txt', 'r') as f:
            for line in f:
                inner_list = [float(elt.strip()) for elt in line.split('\t')]
                list_of_lists.append(inner_list)
            f1_matrix = np.array(list_of_lists)
            for i in range(5):
                plt.plot(list(range(i,6,1)),f1_matrix[i:,i],'--',label='Task'+str(i)+ ' (EWC)')
plt.xlabel("Step")
plt.ylabel("F1")
plt.legend()
plt.show()
```
